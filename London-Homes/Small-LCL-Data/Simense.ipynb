```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas import DataFrame
import datetime
import csv
import matplotlib.dates as mdates
import time



%matplotlib inline
```

```python
def dataset(name):
    df = pd.read_csv(''+str(name)+'', skiprows=1,delim_whitespace=False,header=None, usecols=[0,1,2,3], names=['id','std','date','data'])
    df['datetime']=pd.to_datetime(df['date'].astype(str) ,format='%Y-%m-%d %H:%M:%S.%f')
    return df
```

```python
df = dataset('LCL-June2015v2_0.csv')
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,5), sharex=True)
ax.xaxis.set_tick_params(direction='in',which='both',width=1,length=5,size=5)
plt.scatter(df['datetime'],df['id'])
plt.show()
```
