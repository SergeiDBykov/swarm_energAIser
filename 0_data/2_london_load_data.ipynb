## functions to load data for London dataset

```python
def read_london_file(filename: str, resample= True) -> pd.DataFrame:
    data_path_london = data_path+'London_orig/'
    df = pd.read_csv(data_path_london+filename, skiprows=1,delim_whitespace=False,header=None, usecols=[0,1,2,3], names=['id','mode','date','power'])
    df['date']=pd.to_datetime(df['date'].astype(str) ,format='%Y-%m-%d %H:%M:%S.%f')
    df['power'] = pd.to_numeric(df['power'], errors='coerce')

    return df


def read_london(file_limit = 10):
    data_path_london = data_path+'London_orig/'

    london_files   = glob.glob(data_path_london+'*.csv')

    dfs = []
    for file in tqdm(london_files[:file_limit]):
        fname = file.split('/')[-1]
        df = read_london_file(fname)
        dfs.append(df)

    df_total = pd.concat(dfs, axis=0)
    

    return df_total
```

# Process data for London, prepare data for google drive folder

```python
df_tot = read_london(file_limit = -1)
df_tot['power'] = pd.to_numeric(df_tot['power'], errors='coerce')

df_tot.head(3)
```

```python
#pivot df_tot to get df with date as index and columns as the different houses
df_pivot_std = pd.pivot_table(df_tot.query('mode=="Std"'), index='date', columns='id', values='power') 
df_pivot_std = df_pivot_std.resample('1H').mean()
df_pivot_std.head(3)
```

```python
df_pivot_std.to_pickle(data_path+'London_drive/london_std.pkl_gz', compression='gzip') 
```

```python
#pivot df_tot to get df with date as index and columns as the different houses
df_pivot_tou = pd.pivot_table(df_tot.query('mode=="ToU"'), index='date', columns='id', values='power') 
df_pivot_tou = df_pivot_tou.resample('1H').mean()
df_pivot_tou.head(3)
```

```python
df_pivot_tou.to_pickle(data_path+'London_drive/london_tou.pkl_gz', compression='gzip') 
 
```
