```python
from scripts.utils import data_path, set_mpl
from scripts.read_data import read_london_file, read_london
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import h5py


%matplotlib inline

set_mpl()
```

```python
df_tot = read_london(file_limit = -1)
df_tot['power'] = pd.to_numeric(df_tot['power'], errors='coerce')
df_tot.head(3)
```

```python
#pivot df_tot to get df with date as index and columns as the different houses
df_pivot_std = pd.pivot_table(df_tot.query('mode=="Std"'), index='date', columns='id', values='power') #command takes 4 min
df_pivot_std
```

```python
df_pivot_std.to_pickle(data_path+'london_std.pkl_gz', compression='gzip') #1 min 30 sec
```

```python
#pivot df_tot to get df with date as index and columns as the different houses
df_pivot_tou = pd.pivot_table(df_tot.query('mode=="ToU"'), index='date', columns='id', values='power') #1 min
df_pivot_tou
```

```python
df_pivot_tou.to_pickle(data_path+'london_tou.pkl_gz', compression='gzip') #22 sec  
```
